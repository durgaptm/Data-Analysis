#### Case: Find the best markets to advertise in for an e-learning platform. Since organizing and conducting a survey will take considerable amount of time and money, the existing survey data is considered. The 2017 New Coders Survey conducted by freeCodeCamp is used for this project. The survey data was cleaned in 'R Studio' and the clean data is used in this part of the project.

For an e-learning platform, it is important to understand the following:
* Percentage of people willing to attend bootcamps
* Are current employees attending bootcamps?
* The age group to target
* Top bootcamps people are recommending the most
* Job role that people are interested
* Average money spend by people on bootcamps/courses
* Country to invest in

Next, to understand the average money spend by the people for learning will give an idea on the opportunity for profit. For this, the average money spend by the people per month needs to be calculated.

```python
fcc_no_nulls['MonthsProgramming'].replace(0, 1, inplace = True) # to replace 0's by 1's to avoid division by 0

# New column for the amount of money spend per month
fcc_no_nulls['MoneySpendPerMonth'] = fcc_no_nulls['MoneySpendLearning']/fcc_no_nulls['MonthsProgramming']

# To keep only rows where money spend per month is not null
fcc_no_nulls = fcc_no_nulls[fcc_no_nulls['MoneySpendPerMonth'].notnull()]

# To remove values where CountryLive is null
fcc_no_nulls = fcc_no_nulls[fcc_no_nulls['CountryLive'].notnull()]
```

```python
fcc_no_nulls['CountryLive'].value_counts().head()
```

```python
# Mean sum of money spend by each student per month for the top 4 countries
fcc_no_nulls.groupby('CountryLive').mean()['MoneySpendPerMonth'][['United States of America','India','United Kingdom','Canada']]
```

The people in India can be seen to spend more than double compared to UK. To understand this further, the data can be visualized.

```python
# To get the top 4 countries of interest
top_4 = fcc_no_nulls[fcc_no_nulls['CountryLive'].str.contains('United States of America|India|United Kingdom|Canada')]

# Box plot to visualize the data
sns.boxplot(x='CountryLive', y='MoneySpendPerMonth',data=top_4)
plt.title('Money Spent Per Month Per Country', fontsize = 16)
plt.ylabel('Money per month (US dollars)')
plt.xlabel('Country')
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada'])
plt.show()
```

Some outliers can be seen in US with money spend per month more than 50000 which seems impossible. The values below 20000 can be isolated to get a clearer picture of the data. 

```python
# Remove values above 20000
top_4 = top_4[top_4['MoneySpendPerMonth'] < 20000]

sns.boxplot(x='CountryLive', y='MoneySpendPerMonth',data=top_4)
plt.title('Money Spent Per Month Per Country', fontsize = 16)
plt.ylabel('Money per month (US dollars)')
plt.xlabel('Country')
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada'])
plt.show()
```

India seems to have a few outliers with money spend per month more than 2500. To get a closer look at these, the values can be isolated.

```python
# Isolate the values more than 2500 in India
india_check = top_4[(top_4['MoneySpendPerMonth'] > 2500) & (top_4['CountryLive'] == 'India')]
india_check
```

It can be seen that these respondents have not attended any bootcamps either to indicate the high amount of money spend per month. So these values can be removed to ensure a better result.

```python
# Remove the outliers from India
top_4 = top_4.drop(india_check.index,axis=0)

sns.boxplot(x='CountryLive', y='MoneySpendPerMonth',data=top_4)
plt.title('Money Spent Per Month Per Country', fontsize = 16)
plt.ylabel('Money per month (US dollars)')
plt.xlabel('Country')
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada'])
plt.show()
```

US also seems to have some outliers with money spend per month more than 2500. To examine these, the values can be isolated.

```python
# Isolate US outliers
us_check = top_4[(top_4['MoneySpendPerMonth'] > 6000) & (top_4['CountryLive'] == 'United States of America')]
us_check
```

Out of the 11, 6 of the respondents have attended bootcamps, which could mean the bootcamps might be expensive which resulted in such large amount of money spend on learning. Also 3 of them have less than 3 months programming experience which means the sum  is still large enough to be removed. The rest of the respondents have not attended any bootcamps so it is not completely possible to justify the huge amount. Thus all these 8 values can be removed.

```python
# Remove the respondents who didn't attendent a bootcamp
us_no_bootcamp = top_4[(top_4['CountryLive'] == 'United States of America') & (top_4['MoneySpendPerMonth'] >= 6000) & 
                       (top_4['AttendedBootcamp'] == 0)]
top_4 = top_4.drop(index=us_no_bootcamp.index,axis = 0)
```

```python
# Remove the respondents who dont have programming experience for less than 3 months 
less_than_3_months = top_4[(top_4['CountryLive'] == 'United States of America') & (top_4['MoneySpendPerMonth'] >= 6000) & 
                           (top_4['MonthsProgramming'] <= 3)]

top_4 = top_4.drop(index=less_than_3_months.index,axis = 0)
```

```python
# Recompute mean sum of money spent by students each month
top_4.groupby('CountryLive').mean()['MoneySpendPerMonth']
```

```python
sns.boxplot(x='CountryLive', y='MoneySpendPerMonth',data=top_4)
plt.title('Money Spent Per Month Per Country', fontsize = 16)
plt.ylabel('Money per month (US dollars)')
plt.xlabel('Country')
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada'])
plt.show()
```

Thus, it seems like US is a good option to advertise in as there seems to be more people interested in coding and also willing to pay an average of 140 per month for courses/bootcamps. 
The second option can be Canada as people are willing to spend up 103 per month compared to India (66) and UK (44).
Although, India shows less amount, but the country could also be considered to advertise in as there are many people interested in coding as well as shows potential.
